```python
import os
import csv
import numpy as np
import pandas as pd
```

```python
index = pd.read_csv('passport-index-matrix.csv',index_col ='Passport')
index
```

```python
rankIndex = index.clip_lower(0)
score = rankIndex.sum(axis=1)

#score.sort_values(by = '0', ascending = False)
rank = score.sort_values(ascending = False)
rank = pd.DataFrame(rank)
rank['Rank'] = np.arange(1, len(score)+1)
#rank
rank.to_csv('OrigRank.csv')
```

```python
#np.repeat(score, len(score), axis=1)
#a= np.tile(score, (len(score),1))
#b = np.transpose(a)
rankIndex.iloc[1,3]
```

```python
for i in range(len(score)):
    for j in range(len(score)):
        rankIndex.iloc[j,i] = rankIndex.iloc[j,i]*score[i]
    
rankIndex
```

```python
rankIndex = rankIndex/100
```

```python
passScore = rankIndex.sum(axis=1)
passScore = passScore.sort_values(ascending = False)
passScore
```

### Update rank and Export to csv

```python
passRank = pd.DataFrame(passScore)
passRank['Rank'] = np.arange(1, len(passScore)+1)
passRank.columns=['Score', 'Rank']
passRank.to_csv('PassportRank1.csv')
```

## Compare ranks 

```python
rankSort = rank.sort_values(by='Passport')
rankSort.columns=['Score', 'Rank']
```

```python
# 
passRankSort = passRank.sort_values(by='Passport')
passRankSort.columns=['New Score', 'New Rank']

# Append old rankings and scores in the new file
passRankSort['Old Score']=rankSort['Score']
passRankSort['Old Rank']=rankSort['Rank']
#passRankSort.append(rankSort, ignore_index=True)

# Create update column showing change in Rank
passRankSort['Update'] = passRankSort['Old Rank'] - passRankSort['New Rank']
```

```python
# Save to csv file
passRankSort.to_csv('PassportRank_update.csv')
```
